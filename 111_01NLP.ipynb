```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
import numpy as np

# Sample data: A 10x10 attention matrix evolving over 20 "steps"
data = [np.random.rand(10, 10) for _ in range(20)]

fig, ax = plt.subplots()
im = ax.imshow(data[0], cmap='viridis')

def update(frame):
    im.set_array(data[frame])
    return [im]

ani = animation.FuncAnimation(fig, update, frames=20, interval=200)
plt.close() # Prevents extra plot from showing
HTML(ani.to_jshtml()) # Renders the animation in the Colab cell
```
